```python
import hashlib

def hash_file(filename):
    with open(filename, "rb") as file:
        content = file.read()
        return hashlib.sha256(content).hexdigest()

file_name = "sample_file.txt"
file_hash = hash_file(file_name)

print("SHA-256 File Hash:", file_hash)
```

```python
import hashlib

def hash_file(filename):
    with open(filename, "rb") as file:
        content = file.read()
        return hashlib.sha256(content).hexdigest()

file_name = "sample_file.txt"
file_hash = hash_file(file_name)

print("SHA-256 File Hash:", file_hash)
```

## Lab 2.2 Reflection

**Did the hash value change?**
Yes, the hash completely changed even though only one word was modified.

**Why does even a small file change create a different hash?**
SHA-256 is designed so that any tiny change in input produces a totally different output. This is called the "avalanche effect."

**Why is hashing useful for file integrity checks?**
It lets you verify that a file has not been tampered with. If the hash matches the original, the file is unchanged.
