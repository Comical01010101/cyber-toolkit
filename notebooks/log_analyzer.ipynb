```python
import pandas as pd

logs = pd.read_csv("mock_logs.csv")
logs
```

```python
print("IP Address Frequency:")
print(logs["ip_address"].value_counts())
```

```python
failed_logs = logs[logs["status"] == "FAILED"]

print("Failed Login Attempts:")
print(failed_logs)
```

```python
suspicious_ips = failed_logs["ip_address"].value_counts()
suspicious_ips = suspicious_ips[suspicious_ips > 2]

print("Suspicious IP Addresses:")
print(suspicious_ips)
```

## Lab 2.3 Reflection

**Why are repeated failed login attempts suspicious?**
They may indicate a brute force attack where someone is repeatedly trying to guess a password.

**What might this pattern suggest?**
An attacker could be systematically trying different passwords to gain unauthorized access.

**Which IP addresses appear most suspicious?**
192.168.1.10 (3 failed attempts) and 192.168.1.44 (3 failed attempts) are the most suspicious.

**How could a real security team use this type of analysis?**
They could automate this detection to block IPs that exceed a failure threshold, or trigger an alert for a security analyst to investigate.

## Lab 2.4 Reflection

**Why is logging useful in cybersecurity tools?**
Logs create a record of what happened, when it happened and in what order, essential for investigations.

**What kinds of events should security tools log?**
Login attempts, file access, errors, configuration changes and any detected threats.

**How could logs help during an investigation?**
They provide a timeline of events that investigators can follow to understand how an attack occurred or what data was accessed.
