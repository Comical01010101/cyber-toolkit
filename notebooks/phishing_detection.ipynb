## Phishing Analysis

### Message A
"Your account has been locked. Click here immediately to verify your password: http://secure-login-now.com"

### Message B
"Hi, just checking in about tomorrow's meeting. Let me know if you're available."

### Message C
"You have won a $500 gift card! Claim now by entering your personal details."

## My Phishing Analysis

**Which messages are phishing attempts?**
Message A and Message C are phishing attempts. Message B is legitimate.

**Signs in Message A:**
- Creates urgency ("immediately")
- Suspicious link (http://secure-login-now.com — not an official domain)
- Requests password verification

**Signs in Message C:**
- Too good to be true offer ($500 gift card)
- Requests personal details
- No sender context or official branding

```python
message = input("Enter a message: ")

keywords = ["urgent", "verify", "click", "password", "win", "claim"]

for word in keywords:
    if word in message.lower():
        print("⚠️ Potential phishing message detected")
        break
```

## Lab 3.2 — AI in Cybersecurity

### How can AI help detect cyber threats?
AI can analyze huge amounts of network traffic and log data in real time, flagging unusual patterns much faster than a human analyst could manually review.

### How could AI analyze large amounts of log data?
AI models can be trained to recognize what "normal" activity looks like, and automatically alert when something deviates — for example, a login from an unusual country or at 3am.

### How might AI identify phishing emails?
AI can scan email content for suspicious keywords, check links against known malicious domains, and analyze writing patterns that differ from legitimate messages.

### What happens if AI makes a mistake?
A false positive could block a legitimate user. A false negative could allow a real attack to go undetected. Both have serious consequences in cybersecurity.

### Could attackers use AI too? How?
Yes. Attackers can use AI to write more convincing phishing emails, find software vulnerabilities faster, or automate large-scale attacks.

### Why is human oversight still important?
AI lacks judgment and context. A human analyst can understand nuance, verify edge cases, and make ethical decisions that AI cannot reliably handle on its own.

### What does "Responsible AI" mean in cybersecurity?
Responsible AI in cybersecurity means using AI tools transparently and ethically, with humans remaining in control of critical decisions.It means regularly testing AI systems for bias or errors and ensuring that AI assisted tools are not used to invade privacy or target individuals unfairly.

```python
from sklearn.ensemble import IsolationForest
import numpy as np

# Sample login attempt data
data = np.array([[1], [2], [2], [3], [100]])

model = IsolationForest(contamination=0.2)
model.fit(data)

predictions = model.predict(data)

print(predictions)
```

## Lab 3.3 Reflection

**Which value is suspicious?**
The value 100 is flagged as suspicious (prediction = -1).

**Why is it flagged?**
All other values (1, 2, 2, 3) are clustered closely together. The value 100 is a major outlier — far outside the normal range.

**How could this apply to cybersecurity logs?**
In a real scenario, this could represent a user making 100 login attempts in a short time, while normal users make only 1–3. The AI would flag this as a potential brute-force attack.
